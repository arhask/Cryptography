```python
"""
1.12 part e
Updated version of myExtended2.
Returns (g,u,v) where u is the smallest positive value that satisfies au + bv = gcd(a, b).
"""
def myExtended3(a,b):
    u = 1
    g = a
    x = 0
    y = b
    
    if b == 0: #if b is zero
        v= 0
        return (g,u,v)
        
    while y != 0:
        q = g // y
        t = g % y
        s = u - q * x
        u = x
        g = y
        x = s
        y = t

    if y == 0:
        v = (g - a * u)/ b
        
        u0 = u
        v0 = v
        k = 1
        while u < 0: #if u is negative
            u = u0 + k * b / g 
            v = v0 - k * a / g
            k = k + 1
        return (g,u,v)
```
