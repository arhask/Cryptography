```python
""" 
This algorithm is low-storage variant of the square and-
multiply algorithm
Returns the value g^A (mod N).
"""
def sAm(N,g,A):
    a = g 
    b = 1
    while A > 0:
        if A % 2 == 1:
            b = (b * a) % N
        a = (a * a) % N
        A = floor(A // 2 )
        
    return b
```
